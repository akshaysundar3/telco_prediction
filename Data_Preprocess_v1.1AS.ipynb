```python
# Load Libraries and Dataset

import pandas as pd
import numpy as np

telco = pd.read_table("Telco_data.csv", sep=",")
```

```python
# Check columns in Data

print(telco)
```

```python
# Preprocessing by converting the qualitative to quantitative

telco_proc = telco

telco_proc["gender_proc"] = np.where(telco_proc["gender"]=="Male", 0, 1)
telco_proc["partner_proc"] = np.where(telco_proc["Partner"]=="No", 0, 1)
telco_proc["dependents_proc"] = np.where(telco_proc["Dependents"]=="No", 0, 1)
telco_proc["phone_service_proc"] = np.where(telco_proc["PhoneService"]=="No", 0, 1)
telco_proc["churn_proc"] = np.where(telco_proc["Churn"]=="No", 0, 1)
telco_proc["paperless_billing_proc"] = np.where(telco_proc["PaperlessBilling"]=="No", 0, 1)

telco_proc["multiple_lines_proc"] = np.where(telco_proc["MultipleLines"]=="No", 0, -1)
telco_proc["multiple_lines_proc"] = np.where(telco_proc["MultipleLines"]=="Yes", 1, -1)

telco_proc["internet_service_proc"] = np.where(telco_proc["InternetService"]=="No", 0, 1)
telco_proc["internet_service_proc"] = np.where(telco_proc["InternetService"]=="Fiber optic", 2, 1)

telco_proc["online_security_proc"] = np.where(telco_proc["OnlineSecurity"]=="No", 0, -1)
telco_proc["online_security_proc"] = np.where(telco_proc["OnlineSecurity"]=="Yes", 1, -1)

telco_proc["online_backup_proc"] = np.where(telco_proc["OnlineBackup"]=="No", 0, -1)
telco_proc["online_backup_proc"] = np.where(telco_proc["OnlineBackup"]=="Yes", 1, -1)

telco_proc["device_protection_proc"] = np.where(telco_proc["DeviceProtection"]=="No", 0, -1)
telco_proc["device_protection_proc"] = np.where(telco_proc["DeviceProtection"]=="Yes", 1, -1)

telco_proc["tech_support_proc"] = np.where(telco_proc["TechSupport"]=="No", 0, -1)
telco_proc["tech_support_proc"] = np.where(telco_proc["TechSupport"]=="Yes", 1, -1)

telco_proc["streaming_tv_proc"] = np.where(telco_proc["StreamingTV"]=="No", 0, -1)
telco_proc["streaming_tv_proc"] = np.where(telco_proc["StreamingTV"]=="Yes", 1, -1)

telco_proc["streaming_movies_proc"] = np.where(telco_proc["StreamingMovies"]=="No", 0, -1)
telco_proc["streaming_movies_proc"] = np.where(telco_proc["StreamingMovies"]=="Yes", 1, -1)


#Define a dictionary for Contract

contract_dic = { 'Month-to-month':1, 'One year':2, 'Two year':3}

#Define a dictionary for Payment Method

payment_dic = { 'Electronic check':1, 'Mailed check':2, 'Bank transfer (automatic)':3,
 'Credit card (automatic)':4
}

telco_proc["contract_proc"] = telco_proc["Contract"].map(lambda x: contract_dic[x])
telco_proc["payment_method_proc"] = telco_proc["PaymentMethod"].map(lambda x: payment_dic[x])

telco_proc = telco_proc[["customerID", "gender_proc", "partner_proc", "dependents_proc", "phone_service_proc", 
                        "churn_proc", "paperless_billing_proc", "multiple_lines_proc", "internet_service_proc", 
                        "online_security_proc", "device_protection_proc", "tech_support_proc", "streaming_tv_proc",
                        "streaming_movies_proc", "tenure", "contract_proc", "payment_method_proc", "MonthlyCharges", 
                        "TotalCharges", "SeniorCitizen", "online_backup_proc"]]
print(telco_proc)
```

```python
# Error free Conversion of String to Float

def float_error(x):
    try:
        float(x)
        return(float(x))
    except:
        return 0
    

# Convert Total Charges

telco_proc.TotalCharges = telco_proc.TotalCharges.map(lambda x: float_error(x))

print(telco_proc.TotalCharges)
```

```python
telco_proc.to_csv("Telco_data_preproc.csv")
```
